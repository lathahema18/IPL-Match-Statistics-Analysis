```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matches = pd.read_csv("matches (1).csv")
deliveries = pd.read_csv("deliveries (2).csv")

print(matches.head())
print(deliveries.head())
```

```python
print(matches.isnull().sum())

matches.drop_duplicates(inplace=True)

matches['winner'] = matches['winner'].fillna("No Result")

print("Cleaning Completed")
```

```python
team_wins = matches['winner'].value_counts()

plt.figure(figsize=(10,5))

sns.barplot(
    x=team_wins.values,
    y=team_wins.index
)

plt.title("Most Successful Teams")
plt.xlabel("Wins")
plt.ylabel("Teams")

plt.show()
```

```python
highest_score = deliveries.groupby(
    'batting_team'
)['total_runs'].sum()

highest_score = highest_score.sort_values(
    ascending=False
)

plt.figure(figsize=(10,5))

sns.barplot(
    x=highest_score.values,
    y=highest_score.index
)

plt.title("Highest Team Scores")

plt.show()
```

```python
wickets = deliveries[
    deliveries['dismissal_kind'].notna()
]

top_bowlers = wickets['bowler'].value_counts().head(10)

plt.figure(figsize=(10,5))

sns.barplot(
    x=top_bowlers.values,
    y=top_bowlers.index
)

plt.title("Top Wicket Takers")

plt.show()
```

```python
year_winners = matches.groupby(
    'season'
)['winner'].agg(
    lambda x:x.value_counts().index[0]
)

print(year_winners)
```

```python
latest_years = sorted(
    matches['season'].unique()
)[-5:]

last5 = matches[
    matches['season'].isin(latest_years)
]

print(last5['winner'].value_counts())
```

```python
numeric_data = matches.select_dtypes(
    include=['int64','float64']
)

plt.figure(figsize=(10,8))

sns.heatmap(
    numeric_data.corr(),
    annot=True
)

plt.title(
    "IPL Correlation Heatmap"
)

plt.show()
```
